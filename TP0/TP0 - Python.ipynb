####**Tarea**: iterar elementos de un vector:

####**Tarea**: realizar operaciones matemáticas sobre vectores.

```
import math

x = [3, 4, 2.1, 1.9]
print(x)

s = sum(x)
print("Suma de elementos: {0}".format(s))

min_x = min(x)
max_x = max(x)
print("Mínimo y máximo del vector: {0}, {1}".format(min_x, max_x))

norma = math.sqrt(sum([e**2 for e in x])) # Raiz cuadrada de la suma del cuadrado de los elementos del vector
print("Norma del vector: {0}".format(norma))
```

####**Tarea**: generar un vector con n elementos generados al azar.

```
import random

def aleatorio_1(n):
    """Método 1: dinámico usando bucle for
    """
    v = []
    for i in range(n):
      v.append(random.random())
    return v

def aleatorio_2(n):
    """Método 2:  inicializando e iterando
    """
    v = [0]*n
    print(v)
    for i in range(len(v)):
      v[i] = random.random() # Reemplazo con el valor aleatorio
    return v

def aleatorio_3(n):
    """Método 2: list comprehensions
    """
    return [random.random() for i in range(n)]

n = 6
print('Método 1 (con for dinámico)')
print(aleatorio_1(n))
print('Método 2 (con inicializacion)')
print(aleatorio_2(n))
print('Método 3 (list comprehensions)')
print(aleatorio_3(n))
```

####**Tarea**: definir una función que tome un vector como entrada, y devuelva un valor real como salida.

```
def funcion(x_vector):
    """Devuelve el valor de la función f(x1,x2,...,xn) = x1^2 + x2^2 + ... + xn^2
    """
    return sum([x**2 for x in x_vector])

x = [5, ]
print("f({0}) = {1}".format(x,funcion(x)))

x = [3, 2]
print("f({0}) = {1}".format(x,funcion(x)))

x = [2, 1, -1]
print("f({0}) = {1}".format(x,funcion(x)))
```

## Ejercicio 2
Resolver los siguientes ejercicios de programación.

1.    Implementar un programa que iterativamente genere numeros aleatorios enteros dentro del intervalo [0, 20] y que termine la ejecución al obtener un 0. Debe imprimir en pantalla la cantidad de iteraciones realizadas al terminar la ejecución.

2.    Generar un vector con n elementos binarios (valor 0 o 1 generado al azar). Luego contar cuantos ceros hay en la primera mitad del vector (posición 0 a n/2) y cuantos unos hay en la segunda mitad del vector. Probar con n = 10 y n = 13.


```
'''Ejercicio 1: Implementar un programa que iterativamente genere numeros aleatorios enteros dentro del intervalo [0, 20] y que termine la ejecución
al obtener un 0. Debe imprimir en pantalla la cantidad de iteraciones realizadas al terminar la ejecución.'''

import random

contador = 0
numero = -1   # Numero distinto del rango para entrar al bucle

while numero != 0:
    numero = random.randint(0, 20) # Usamos esta funcion de la libreria random porque me devuelve valores enteros al azar dentro del rango intervalo cerrado
    contador += 1
    print(numero)

print("\nCantidad de iteraciones realizadas hasta obtener 0:", contador)
```

```
'''Generar un vector con n elementos binarios (valor 0 o 1 generado al azar). Luego contar cuantos ceros hay en la primera mitad del vector (posición 0 a n/2)
y cuantos unos hay en la segunda mitad del vector. Probar con n = 10 y n = 13.'''

import random

def crear_vector(n):
  v = []
  for i in range(n):
    v.append(random.randint(0, 1))
  return v

def contador_ceros(v):
  contador = 0
  primera_mitad = v[:len(v)//2]
  for i in primera_mitad:
    if i == 0:
      contador += 1
  return contador

def contador_unos(v):
  contador = 0
  segunda_mitad = v[len(v)//2:]
  for i in segunda_mitad:
    if i == 1:
      contador += 1
  return contador

n = int(input("Ingrese el tamaño del vector: "))
vector = crear_vector(n)
print("\tEl vector creado es:", vector)
print("\tCantidad de 0 en la primera mitad del vector:", contador_ceros(vector))
print("\tCantidad de 1 en la segunda mitad del vector:", contador_unos(vector))

n = 10
vector = crear_vector(n)
print("\nPara n = 10")
print("\tEl vector creado es:", vector)
print("\tCantidad de 0 en la primera mitad del vector:", contador_ceros(vector))
print("\tCantidad de 1 en la segunda mitad del vector:", contador_unos(vector))

n = 13
vector = crear_vector(n)
print("\nPara n = 13")
print("\tEl vector creado es:", vector)
print("\tCantidad de 0 en la primera mitad del vector:", contador_ceros(vector))
print("\tCantidad de 1 en la segunda mitad del vector:", contador_unos(vector))
```
