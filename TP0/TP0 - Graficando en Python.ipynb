## Ejercicio 1
Dada la ecuación de la recta, se solicita emplear la librería Matplotlib (o alguna de su elección) para graficar a través de una función los puntos de la misma.

####**Tarea**: Crear una recta de 20 elementos.

```
print('Configurar parámetros de la recta.')
cant_elementos = 20
a = 2
b = 10
recta = [a*x+b for x in range(cant_elementos)]

print(f"Valores de la recta: {recta}")
```

####**Tarea**: Usar la librería de Matplotlib para graficar la recta.

```
import matplotlib.pyplot as plt
%matplotlib inline

x_values_recta = [i for i in range(cant_elementos)]
plt.plot(x_values_recta,recta,'ro-',label=f'recta simple a={a} - b={b}')
plt.xlabel('Valres de X')
plt.ylabel('Valres de Y')
plt.title(f"Recta en el plano: pendiente {a} y ordenada al origin {b}.")
```

####**Tarea**: Usando de base el código anterior, graficar una parábola con la configuración que desee.

## Ejercicio 2
Graficar la norma del vector para una corrida con 10, 50, 100, y 500 puntos aleatorios. De ser necesario, buscar lo que se denomina "norma de un vector".

```
import random
import matplotlib.pyplot as plt
import math

puntos = [10,50,100,500]
normas = []

for i in puntos:
  vector = [random.random() for x in range(i)]
  norma = math.sqrt(sum([e**2 for e in vector]))
  normas.append(norma)

for i, n in zip(puntos,normas):
  print(f"Vector de: {i} elementos - Norma del vector: {n}")

plt.figure(figsize=(10,6))
plt.plot(puntos,normas,marker='o',color='purple',label="Norma del vector")
plt.xlabel("Cantidad de puntos")
plt.ylabel("Norma del vector")
plt.title("Norma del vector")
plt.grid(True)
plt.legend()
plt.show()
```

El programa genera vectores de distinta longitud con valores aleatorios, calcula su norma euclídea y guarda esos resultados en una lista para luego representarlos gráficamente. Se observa que, al aumentar la cantidad de elementos del vector, la norma también crece, ya que refleja la magnitud total de los componentes. De esta forma, el código no solo automatiza el cálculo, sino que también permite visualizar la relación entre el tamaño del vector y su norma.
