El siguiente código es para demostrar que no es adecuado usar "round" para generar numeros enteros aleatorios.

```
import random
import math
import matplotlib.pyplot as plt

#CONSIGNA:
#generar N numeros ENTEROS aleatorios con probabilidad uniforme, entre a y b

def generar1(N, a, b):
  return [round(random.random()*(b-a)+a) for i in range(N)]

def generar2(N, a, b):
  return [int(random.random()*(1+b-a)+a) for i in range(N)]

def generar3(N, a, b):
  return [math.ceil(random.random()*(1+b-a)+a-1) for i in range(N)]

def generar4(N, a, b):
  return [random.randint(a, b) for i in range(N)]

#contar las veces que sale cada numero (cuantos numeros 0, cuantos 1, etc.)
def contar(vector, texto):
  n = max(vector)+1
  cantidad = [0]*n
  for i in vector:
      cantidad[i] = cantidad[i] + 1

  plt.bar(range(n), cantidad)
  plt.title(texto)
  plt.xlabel("Número")
  plt.ylabel("Cantidad de veces")
  plt.show()
```

Al generar con método con "round" la probabilidad no es uniforme (los extremos tienen una probabilidad menor). Con "int", en cambio, sí se logra una probabilidad uniforme.

```
N = 10000

vector = generar1(N, 0, 10)
contar(vector, 'Cantidad de veces que sale cada numero, usando "rand" (ERROR)')

vector = generar2(N, -4, 10)
contar(vector, 'Cantidad de veces que sale cada numero, usando "int" (BIEN)')

vector = generar3(N, 0, 10)
contar(vector, 'Cantidad de veces que sale cada numero, usando "ceil" (BIEN)')

vector = generar4(N, 0, 10)
contar(vector, 'Cantidad de veces que sale cada numero, usando "randint" (BIEN)')
```

Al realizar las pruebas del programa donde todos los generadores están calibrados con el mismo rango (0 a 10) tuvimos los siguientes resultados:

* Para 100 tiros hasta 500 tiros  notamos una irregularidad en la frecuencia de generación de algunos valores provocando que la distribución de los valores no sea uniforme. Esto pasa incluso en los que son considerados buenos generadores.


* Para 1000 tiros vemos que con round la distribución de aparición de los extremos inferior y superior tienden a generarse con menor frecuencia, a un 50% menos que los valores que se encuentran dentro del rango, podemos decir qué para el caso de round los valores tienden a seguir una distribución normal con lo cual los valores más cercanos a la media son generados con mayor frecuencia qué los valores extremos.



En cambio, para los demás generadores, la distribución puede variar entre ejecuciones, pero a medida que aumenta la cantidad de intentos, la frecuencia de aparición de los valores tiende a estabilizarse, acercándose a una distribución uniforme.

Para detectar el error en el generador round() la cantidad mínima de tiros es a partir de los 500 tiros, ya que se nota que los valores extremos aparecen con menor frecuencia. Con 10000 tiros el error es claro y es estadísticamente significativo.

Ahora para los otros generadores, con 100 tiros hay mucha variabilidad en la frecuencia pero probando con 1000 e incluso 10000 tiros podemos confirmar que los generadores son uniformes.

Para la **función round()** se investigó qué en ciertos casos presenta un error de redondeo debido a la representación de los datos de punto flotante en binario. De acuerdo a lo qué se investigó cualquier número qué no se pueda construir cómo potencias de dos no puede representarse exactamente cómo un float, por lo cual se necesita aproximarlo, en ocasiones la aproximación más cercana será menor qué el número en cuestión.

En el caso de la **función int()** lo qué hace es castear (conversión de tipos) a valores enteros, para los valores de punto flotante lo qué hace está función es directamente truncarlos, trunca los valores hacia cero, por ejemplo: si aplicamos int(3.45) siempre nos devolverá el valor a la izquierda del punto (la parte entera) independientemente de las cifras luego del punto, en el caso de un negativo cómo -5.40 nos dará -5, o si fuese -5.99 nos dará -5 también.

En el caso de **math.ceil()** lo qué hace está función es redondear hacia arriba el valor para cualquier caso, ejemplo: si aplicamos math.ceil() a 10.1 este nos va a devolver 11, para 10.01 también lo aplica, si se tiene 10.000001 también redondea hacia arriba.

Con **random.randint()** se utiliza directamente el generador de números aleatorios de python, implementa en su lógica el casteo int() y el cálculo del rango de valores internamente de forma tal qué trata de asegurar una distribución normal de los valores dentro del rango. En los métodos anteriores se notó el uso del cálculo del rango en conjunto con el uso del método para reducir el sesgo, en cambio random.randint() implementa internamente el cálculo del rango y el redondeo.

En conclusión, los problemas observados en la generación de números aleatorios se deben a la forma en que cada método transforma los números decimales en enteros. En particular, métodos como round() introducen un sesgo que reduce la probabilidad de los valores extremos, mientras que otros como int(), ceil() o randint() logran distribuciones más uniformes.
